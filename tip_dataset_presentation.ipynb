```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
print(tips.head(10))
print(tips['day'].value_counts())
print(tips.info())
```

```python
tips.head(15)
tips['tip_ratio']=100*tips['tip']/tips['total_bill']
```

```python
sns.scatterplot(tips['tip'],tips['total_bill'])
plt.show()
```

```python
sns.scatterplot(tips['total_bill'],tips['tip_ratio'])
plt.show()
sns.regplot(x="total_bill", y="tip_ratio", data=tips)
plt.show()
```

```python
plt.figure(figsize=(6,7))

ax=sns.lineplot(x='size',y='tip_ratio', data= tips, color="coral", label="line")
ax.set(xlabel = "Party Size", ylabel = "Tip Ratio", title = "Tip ratio by Party Size")
ax.legend_.remove()
plt.show()
```

```python
grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist,"tip_ratio", bins=np.linspace(0, 40, 15),edgecolor="k")
grid.set(xlabel = "Count", ylabel = "tip ratio")
plt.show()
```

```python
grid = sns.FacetGrid(tips, row="smoker", col="time", margin_titles=True)
grid.map(plt.hist,"tip_ratio", bins=np.linspace(0, 40, 15),edgecolor="k")
grid.set(xlabel = "Count", ylabel = "tip ratio")
plt.show()
```

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot("day", "total_bill", "sex", data=tips, kind="box")
    g.set_axis_labels("Day", "total bill")
plt.show()

sns.catplot(x="day", y="tip",  kind="swarm", data=tips)
plt.show()
sns.boxplot(x="day", y="tip", data=tips)
plt.show()
```

```python
sns.boxplot(x = 'day', y= 'total_bill', data = tips)
plt.show()
```

```python
sns.countplot(x='day',hue='smoker',data = tips)
plt.show()
```

```python
sns.catplot(x="day", y="tip_ratio", hue="time",
            kind="violin", data=tips)
plt.show()
```

```python
# Pass in the bins parameter to adjust the bin sizes for the histogram
sns.distplot(tips['total_bill'], kde=False, bins=50)
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
```

```python
barplot = sns.barplot(x = "day", y = "tip_ratio", data = tips, order = ["Thur", "Fri", "Sat", "Sun"])
barplot.set(xlabel = "Day", ylabel = "tip ratio", title = "tip ratio by Day")
plt.show()
```

```python
sns.boxplot(x='day', y='tip_ratio', data=tips, palette='Blues',order = ["Thur", "Fri", "Sat", "Sun"])
plt.show()
```

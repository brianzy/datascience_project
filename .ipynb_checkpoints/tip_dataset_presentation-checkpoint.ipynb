```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
print(tips.head(10))
print(tips.describe())
print(tips.info())
```

```python
# Pass in the bins parameter to adjust the bin sizes for the histogram
sns.distplot(tips['total_bill'], kde=False, bins=50)
plt.show()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
plt.show()
```

```python
barplot = sns.barplot(x = "day", y = "total_bill", data = tips, order = ["Thur", "Fri", "Sat", "Sun"])
barplot.set(xlabel = "Day", ylabel = "Total Bill", title = "Total Bill by Day")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips, palette='Blues',order = ["Thur", "Fri", "Sat", "Sun"])
plt.show()
```

```python

plt.figure(figsize=(5,4))
sns.boxplot(x='sex', y='tip', data=tips, palette='Blues')

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.show()
plt.figure(figsize=(7,5))

sns.violinplot(x="day", y="tip", hue="sex",
                     data=tips, palette="Set2", split=True,
                     scale="count", inner="stick",
                     scale_hue=False, bw=.2)
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", hue="smoker",
                    data=tips, palette="muted",split=True)
plt.show()
```

```python
sns.violinplot(x="day", y="tip", hue="smoker",
                    data=tips, palette="muted",split=True)
plt.show()
```

```python
sns.catplot(x="sex", y="total_bill",
                hue="smoker", col="time",
               data=tips, kind="violin", split=True,
                height=4, aspect=.7)
plt.show()
```
